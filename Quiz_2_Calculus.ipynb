# Task 1: Write a dynamic function to find the derivative of any function f(x)


```python
"""
Importing necessary modules, matplotlib for plotting diagrams, sympy for
symbolic mathematics, makes it possible to calculate the derivative.
"""
from sympy import diff, Symbol, parse_expr
from sympy.solvers import solve
import numpy as np
import matplotlib.pyplot as plt
```

```python
# A function that does a derivative for any function
def dy_dx(f):
  """
  Args:
  f : Function to differentiate.
  Returns:
  Returns the derivative of a function with respect to x
  """
  return diff(f)
```

**Task 4: Identifying Minima and Maxima**
Creating separate arrays to store x values of minima and maxima of the function.
Print the arrays to display the critical points of the function

```python
# Creating separate arrays to store maximas and minimas
arr_minimas = []
arr_maximas = []

# Creates a symbol x to represent the variable x
x = Symbol('x')

# Calculates the first derivative of f(x) = f'(x)
f_prime = dy_dx(f(x))

# Then calculate the second derivative of f(x) = f''(x)
f_prime_prime = dy_dx(f_prime)

# Calculates the stationary_points using sympy solve
stationary_points = solve(f_prime)

# Now classify each point as either a minima or a maxima
for point in stationary_points:

  # Calculate the value of the second_derivative at a point
  second_derivative = f_prime_prime.evalf(subs={x: point})

  # If the second_derivative is greater than zero(positive), the point is a minima
  if second_derivative > 0:
    arr_minimas.append(point)

  # If the second_derivative is less than zero(negative value), the point is a maxima
  if second_derivative < 0:
    arr_maximas.append(point)

# Printing out the minimas and maximas
print(f'the array of minimas {arr_minimas} and the array of maximas is {arr_maximas}')
```

**Task 5: Computing the Global Minima and maxima**
Determine the global minima and maxima of the quadratic function.
Print the coordinates of these points to identify the overall minimum and maximum values of the function.


```python
# A global maxima is the highest y intercept while the global minima is the lowest y intercept

# Finding the global maxima
# First initialise the highest y value  as -infinity
highest_y = float('-inf')

# Initialise the global_maximum to None
global_maximum = None

# Loop through the array of maximas
for point in arr_maximas:

  # Calculate the y value of each maxima
  y = f(point)

  # If it exceeds the highest_y seen so far
  if y > highest_y:

    # That becomes our new global_maximum
    global_max_point = point
    # And we update the highest_y seen so far
    highest_y = y

# Finding the global minimum
# Initialise the lowest y value so far as -infinity
lowest_y = float('inf')

# Initialise the global_minimum to None
global_minimum = None

# Loop through the array of minimas
for point in arr_minimas:

  # Calculate the y value of each minima
  y = f(point)

  # If it's lower than the lowest_y seen so far
  if y < lowest_y:

    # That becomes our new global_minimum
    global_min_point = point

    # And we update the lowest_y seen so far
    lowest_y = y

global_maximum = (global_max_point,highest_y)
global_minimum = (global_min_point,lowest_y )

# Printing out the global maxima and minima
print(f'The global maxima is {global_maximum}')
print(f'The global minima is {global_minimum}')
```
